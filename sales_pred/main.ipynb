```python
from sklearn.feature_selection import VarianceThreshold

from sklearn.metrics import r2_score

# Instancie o seletor de características
seletor = VarianceThreshold(threshold=0.1)  # Escolha o limite de variância desejado

# Aplique o seletor aos seus dados de treinamento
X_train = seletor.fit_transform(X_train)

# Aplique o mesmo seletor aos seus dados de teste
X_test = seletor.transform(X_test)

# Inicialize o modelo RandomForestRegressor
random_forest = RandomForestRegressor(max_depth=None, min_samples_leaf=4, min_samples_split=5, n_estimators=100)

# Treine o modelo com os dados de treinamento selecionados
random_forest.fit(X_train, y_train)

# Faça previsões nos dados de teste selecionados
previsoes = random_forest.predict(X_test)

# Calcule o coeficiente de determinação (R²)
r2 = r2_score(y_test, previsoes)

print(f'Coeficiente de Determinação (R²) nos dados de teste após a seleção de características: {r2}')
```

```python
import matplotlib.pyplot as plt

# Fazer previsões nos dados de treinamento para visualização
previsoes_treinamento = random_forest.predict(X_train)

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.scatter(y_train, previsoes_treinamento, color='blue', label='Previsões nos dados de treinamento')
plt.scatter(y_test, previsoes, color='red', label='Previsões nos dados de teste')
plt.xlabel('Valores Reais (y)')
plt.ylabel('Previsões (y_pred)')
plt.title('Comparação entre Valores Reais e Previsões')
plt.legend()
plt.grid(True)
plt.show()
```
